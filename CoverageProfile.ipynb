```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.pyplot import *
from __future__ import print_function
from collections import defaultdict
import numpy as np
from pylab import *
import matplotlib.ticker as mticker
```

```python
cov={}
count={}
cov=[0]*1501  
count=[0]*1501  
count0 =0
i =0 
summ =0 
for line in open("mpg_L14489-1_Tobacco-lib-RG-12-CTTGTA_S7_R1_001.coverage.out"):
    if line.startswith('mpg'):
        break
    x = line.rstrip().split(' ')
    cov[i]=float(x[0])
    count[i]=float(x[1])  
    summ += count[i]
    i +=1
    if i >1500:
        break


 
#for i in range(0,1500):
#         count[i] =(count[i]/summ)
print(summ)


 

Y =count  
 
X =cov 

# Compute the CDF
CY = np.cumsum(Y)
plt.rcParams["figure.figsize"] = (11,5)
fig, ax = plt.subplots(1,1)

# Plot both
plot(X,CY,color='#4863A0', linestyle='solid', linewidth=4, label ='Coverage')
#plot(X,CY,'r--')
#plt.plot(X, CY,color='#4863A0', linestyle='solid', linewidth=4, label ='Mapping Coverage')
plt.title('Coverage profile')

#plt.axis(xmax=7000.0)
plt.axis(xmin=0)
#plt.axis(ymax=3e+09)
ax.set_yticks(ax.get_yticks()[::2])
plt.xlabel('Coverage')
#plt.ylabel('Cummulative bases count')
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.2e'))

plt.show()
print('done')
```
