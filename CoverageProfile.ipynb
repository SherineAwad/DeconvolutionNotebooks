```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.pyplot import *
from __future__ import print_function
from collections import defaultdict
import numpy as np
from pylab import *
import matplotlib.ticker as mticker
```

```python
cov= []
count= []
summ =0.0
for line in open("mpg_L14489-1_Tobacco-lib-RG-2-CGATGT_S1_R1_001.coverage.out"):
    if line.startswith('mpg'):
        break
    x = line.rstrip().split(' ')
    cov.append ( float(x[0]) )
    count.append( float(x[1]) ) 
    summ  += float( x[1]  )

fig, ax = plt.subplots(1,1)
plt.rcParams["figure.figsize"] = (7, 4)


#for i in range(0, len(cov)): 
#    count[i] = count[i]/summ 
y = count  
y = np.cumsum(y)
x = cov  
plt.plot(x,y, marker='.', linestyle ='none',linewidth=0.001) 
plt.title('CGATGT_S1 Coverage Profile')
plt.xlabel('Coverage')
plt.ylabel ('Cummulative Distribution')
plt.margins(0.02) 
plt.show()
print('done')
```
