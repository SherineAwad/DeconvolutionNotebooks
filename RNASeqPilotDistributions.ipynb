```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def read_file(infile):
    myfile = open(infile,'r')
    readlen = []
    total = []
    for  row in myfile:
        line = row.split()
        readlen.append(line[1])
        total.append(line[0]) 
    return total, readlen
```

# Contaminated Set

```python
#======Hypersegmented 1 
total, readlen = read_file('SLX-16108.D701rna_D504rna.H3GNNBBXY.s_1.r_1.trimmed.read_length.txt')

x = readlen
y =[]
for i in range(0, len(total)): 
    y.append(float(total[i])/1)
BTtotal, BTreadlen = read_file('SLX-16108.D701rna_D504rna.H3GNNBBXY.s_1.r_1_read_length.txt')

plt.bar(x,y)

for i in range(0, len(BTtotal)): 
    y.append(float(BTtotal[i])/1)
plt.bar(BTreadlen,y)
plt.title('Round1 HS1 Read Length Distribution')
plt.xlabel('Read Length')
plt.ylabel ('Total')
plt.savefig('Round1 HS1.png')
plt.show()
plt.clf()
#=========Hypersegmented 2
total, readlen = read_file('SLX-16108.D702rna_D504rna.H3GNNBBXY.s_1.r_1.trimmed.read_length.txt')

x = readlen
y =[]
for i in range(0, len(total)): 
    y.append(float(total[i])/1)
BTtotal, BTreadlen = read_file('SLX-16108.D702rna_D504rna.H3GNNBBXY.s_1.r_1_read_length.txt')

plt.bar(x,y)

for i in range(0, len(BTtotal)): 
    y.append(float(BTtotal[i])/1)
plt.bar(BTreadlen,y)
plt.title('Round1 HS2 Read Length Distribution')
plt.xlabel('Read Length')
plt.ylabel ('Total')
plt.savefig('Round1 HS2.png')

plt.show()
plt.clf()
 
```
