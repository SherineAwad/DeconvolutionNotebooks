```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def read_file(infile):
    myfile = open(infile,'r')
    readlen = []
    total = []
    for  row in myfile:
        line = row.split()
        readlen.append(line[1])
        total.append(line[0]) 
    return total, readlen
```
