```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.tools as tls
```

```python
myfile = open('SLX-17233.D701rna_D501rna.H3FYGBBXY.s_5.r_1_read_length.txt','r')
readlen = []
total = []
for  row in myfile:
    line = row.split()
    readlen.append(line[1])
    total.append(line[0]) 
y = total 
x = readlen 
 
width = 0.4
plt.bar(x, y, width, color="blue")
plt.title('Reads Length Distribution')
plt.ylabel('Frequency')
plt.xlabel('Read Length')

plt.savefig('701_501.png',dpi = 600) 
plt.show()
 
```
